```python
def threes():
    """Il s'agit de la fonction principale permettant de jouer au jeu Threes"""
    

    action=get_user_menu(None)
    
    while action!="Q":
        
        if action=='N':
            partie=create_new_play(1)
        elif action=='L':
            partie=restore_game()
            
        play=cycle_play(partie)
    
        if not play:
            action=get_user_menu(partie)
            if action=="Q":
                return 
            elif action=='N':
                partie=create_new_play(1)
            elif action=='L':
                partie=restore_game()
            elif action=="S":
                save_game(partie)
            elif action=="C":
                cycle_play(partie)
        else:
            print("Partie Terminée ! Score final : "+str(partie['score']))
        
        action=get_user_menu(partie)
        
    
    
    
```
